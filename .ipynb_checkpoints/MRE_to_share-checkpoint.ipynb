```python
%matplotlib inline
from parcels import FieldSet, ParticleSet, JITParticle, AdvectionRK4_3D,ParticleFile,ErrorCode,AdvectionRK4, Variable,VectorField,Field, FieldSamplingError
from glob import glob
from datetime import timedelta as delta
from datetime import date
import datetime
import xarray as xr
from parcels import plotTrajectoriesFile, GeographicPolar,Geographic
from parcels import logger #, XarrayDecodedFilter
import geopandas as gpd
import pandas as pd
import numpy as np
from shapely.geometry import Point
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

import numpy.ma as ma
from netCDF4 import Dataset
from scipy import interpolate
from scipy.interpolate import RectBivariateSpline
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from matplotlib.colors import ListedColormap
from matplotlib.lines import Line2D
from copy import copy
import cmocean
import pytz
```

## creating new def and functionality to the kernel

```python
def saveUV (particle, fieldset, time):
    (u, v) = fieldset.UV[time, particle.depth, particle.lat, particle.lon]
    particle.U_effective=u*particle.dt
    particle.V_effective=v*particle.dt
    
class GBRParticle(JITParticle):
    U_effective=Variable('U_effective', dtype=np.float32, initial = 0.)
    V_effective=Variable('V_effective', dtype=np.float32, initial = 0.)
    
def DeleteParticle(particle, fieldset, time):
    particle.delete()
```

## Reading files
- For this MRE I'm using only one output frpom the GBR1 model

```python
data_path = '/home/por07g/Documents/Projects/GBR_modeling/GBR_oceanparcels/GBR_main/Data/'
files = sorted(glob(data_path+'gbr1_simple_*.nc'))
print(files)
mesh_mask = data_path + 'coordinates.nc'
## Files
filenames = {'U': {'lon': mesh_mask, 'lat': mesh_mask, 'depth': files[0], 'data': files},
             'V': {'lon': mesh_mask, 'lat': mesh_mask, 'depth': files[0], 'data': files}}

variables = {'U': 'u',
             'V': 'v'}

dimensions = {'U': {'lon': 'longitude', 'lat': 'latitude', 'depth': 'zc', 'time': 'time'},
              'V': {'lon': 'longitude', 'lat': 'latitude', 'depth': 'zc', 'time': 'time'}}
```

## creating the Fielset and seeding the particles

```python
fieldset = FieldSet.from_nemo(filenames, variables, dimensions, allow_time_extrapolation = True)
release_depth = -5.35
npart = 9  # number of particles to be released
lon = np.linspace(146.17, 146.17, int(npart), dtype=np.float32)
lat = np.linspace(-16.85, -16.8, int(npart), dtype=np.float32)
z = np.ones(npart) * release_depth
pset = ParticleSet.from_list(fieldset, pclass=GBRParticle,
                                 lon=lon,
                                 lat=lat,
                                 depth=z)  
```

## Runing OP and creating the output

```python
kernels = pset.Kernel(saveUV) + pset.Kernel(AdvectionRK4) 
output_file = "/home/por07g/Documents/Projects/GBR_modeling/GBR_oceanparcels/GBR_main/simple_run"
pfile = pset.ParticleFile(output_file, outputdt = delta(hours=1))
pset.execute(kernels,
             runtime=delta(days=1),
             dt=delta(hours=1),
             output_file=pfile,
             recovery={ErrorCode.ErrorOutOfBounds: DeleteParticle})
pfile.export()  # export the trajectory data to a netcdf file
```
